## Task-01 Creating a bar chart and histogram to visualise the distribution variables in dataset.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("AmazonSalesData.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'Total Revenue': 'Revenue'}, inplace=True)
```

```python
revenue_by_region = df.groupby('Region')['Revenue'].sum().reset_index()
regions = revenue_by_region['Region']
total_revenue = revenue_by_region['Revenue']
```

```python
fig = plt.figure(figsize=(10, 7))
bars = plt.bar(regions, total_revenue, color='skyblue')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.title('Total Revenue by Region')
plt.xticks(rotation=45, ha='right') 
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
plt.hist(df['Total Profit'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit')
plt.show()
```
